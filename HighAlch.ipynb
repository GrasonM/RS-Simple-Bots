```python
import pyautogui
import cv2
import numpy as np
import time
import random
from win32gui import GetWindowText, GetForegroundWindow
from pynput.keyboard import Key, Controller, Listener
import time as ti
import datetime
from datetime import datetime
import re
import os
import random
import keyboard
import sys

global Active, count, keyboard, current_window, desired_window_name 

# Load images of the High Alchemy spell and the item to be alched
high_alch_img = 'high_alch.png'
item_img = 'item.png'
item_img2 = 'item2.png'
continue_img = 'continue.png'
# Function to find the coordinates of an image on the screen
def find_image(image_path, confidence=0.8):
    try:
        pyautogui.getWindowsWithTitle("RuneScape")[0].maximize()
    except Exception as e:
        print(f"An error occurred: {e}")
    
    screen = pyautogui.screenshot()
    screen = cv2.cvtColor(np.array(screen), cv2.COLOR_RGB2BGR)
    template = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    result = cv2.matchTemplate(cv2.cvtColor(screen, cv2.COLOR_BGR2GRAY), template, cv2.TM_CCOEFF_NORMED)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

    if max_val >= confidence:
        return max_loc
    return None

# Function to perform high alchemy with random delays and error handling
def cast_high_alchemy():
    while True:
        try:
            # Find and click High Alchemy spell
            alch_pos = find_image(high_alch_img)
            #coords = pyautogui.locateOnScreen('benz.png', region=(0,0,1366,768), grayscale=True, confidence=0.5)
            if alch_pos:
                pyautogui.moveTo(alch_pos[0] + 10, alch_pos[1] + 10, duration=0.25)
                pyautogui.click(alch_pos[0] + 10, alch_pos[1] + 10)  # Offset to click the center
                time.sleep(0.75)
            else:
                print("High Alch spell not found.")
                continue

            # Random delay to mimic human behavior
            time.sleep(random.uniform(0.5, 1.0))

            # Find and click the item to be alched
            item_pos = find_image(item_img)
            item_pos2 = find_image(item_img2)
            #coords1 = pyautogui.locateOnScreen('benz.png', region=(0,0,1366,768), grayscale=True, confidence=0.5)
            #coords2 = pyautogui.locateOnScreen('benz.png', region=(0,0,1366,768), grayscale=True, confidence=0.5)
            if item_pos:
                pyautogui.moveTo(item_pos[0] + 10, item_pos[1] + 10, duration=0.25)
                pyautogui.click(item_pos[0] + 10, item_pos[1] + 10)  # Offset to click the center
                time.sleep(0.95)
            else:
                print("Item 1 not found.")
                pass

            if item_pos2:
                pyautogui.moveTo(item_pos2[0] + 10, item_pos2[1] + 10, duration=0.25)
                pyautogui.click(item_pos2[0] + 10, item_pos2[1] + 10)  # Offset to click the center
                time.sleep(0.95)
            else:
                print("Item 2 not found.")
                pass

            # Find and click I wish to continue
            continue_pos = find_image(continue_img)
            #coords3 = pyautogui.locateOnScreen('benz.png', region=(0,0,1366,768), grayscale=True, confidence=0.5)
            if continue_pos:
                pyautogui.moveTo(continue_pos[0], continue_pos[1], duration=0.5)
                pyautogui.click(continue_pos[0], continue_pos[1])
                #pyautogui.click(continue_pos[0] + 10, continue_pos[1] + 10)  # Offset to click the center
            else:
                print("Item not found.")
                continue

            # Random delay to mimic human behavior
            #time.sleep(random.uniform(1.0, 2.0))

        except Exception as e:
            print(f"An error occurred: {e}")
            time.sleep(5)  # Wait before retrying to prevent rapid error repetition

if __name__ == "__main__":
    cast_high_alchemy()
```
